## Les conditions

```python
# si a est supérieur à 1 afficher "a est supérieur à 1"

a = 2
if a > 1:
   print("a est supérieur à 1")
```

```python
# Si a est supérieur à 3 afficher "à supérieur à 3"
# sinon afficher "a est inférieur ou égal à 3"

if a > 3:
    print("a est supérieur à 3")
else:
    print("a est inférieur ou égal à 3")
```

```python
# Si a est égal à 2 afficher "a est égal à 2"

if a == 2:
    print("a est égal à 2")
```

```python
# Condition complexe :
# Si a n'est pas égal à 1
# et si a est inférieur à 3
# afficher a est inférieur à 3

if a != 1 and a < 3:
    print("a est inférieur à 3")
```

```python
# Si a est différent de 1
# ou si a est inférieur 0
# afficher "Bonjour"

if a != 1 or a < 0:
    print("Bonjour")
```

```python
# Si a est égal à 1 afficher "a est égal à 1"
# sinon si a est inférieur à 1 afficher "a est inférieur à 1"
# sinon si a est supérieur à 1 afficher "a est supérieur à 1"

if a == 1:
  print("a est égal à 1")
elif a < 1:
  print("a est inférieur à 1")
else:
  print("a est supérieur à 1")
```

```python
# Condition avec une chaîne de caractère

char = "Bonjour"

if char:
  print("La chaîne de caractère n'est pas vide")
```

```python
# Piège de la condition if sur des valeurs nulles

x = 0

if x:
  print("x a une valeur")
else:
  print("x n'a pas de valeur")
```

## Les listes

```python
# Définir une liste

ma_liste = [10, 20, 40, 20, 5]

print(ma_liste)
```

```python
# Accéder au premier élément de la liste

ma_liste[0]
```

```python
# Accéder au deuxième élément de la liste

ma_liste[1]
```

```python
# Accéder au dernier élément de la liste

ma_liste[-1]
```

```python
# Accéder aux plusieurs éléments en fonction de l'index

ma_liste[1:3]
```

```python
# Modifier un élément d'une liste

ma_liste[0] = 15
print(ma_liste)
```

```python
# Insérer un élément dans une liste

ma_liste.insert(2, 35)
print(ma_liste)
```

```python
# Ajouter un élément à la dernière position

ma_liste.append(20)
print(ma_liste)
```

```python
# Supprimer un élément d'une liste avec pop

ma_liste.pop(2)
print(ma_liste)
```

```python
# Supprimer l'élément 20

ma_liste.remove(20)
print(ma_liste)
```

```python
# Enlever tous les éléments dont la valeur est 20 de la liste
for i in ma_liste:
  if i == 20:
    ma_liste.remove(i)

print(ma_liste)
```

```python
# Ordonner une liste

ma_liste.sort()
print(ma_liste)
```

```python
# Inverser l'ordre des éléments

ma_liste.reverse()
print(ma_liste)
```

```python
# Ordonner une chaine de caractère

char_list = ["Bob", "Alice", "Gargamel"]
char_list.sort()
print(char_list)
```

```python
# Ordonner une chaine de caractère selon un critère

char_list.sort(key=len)
print(char_list)
```

```python
# Ajouter une liste à une liste liste2 existante

ma_liste.extend(char_list)
print(ma_liste)
```

```python
# Référence et copie

ma_copie = ma_liste
print(ma_copie)
```

```python
# Enlever le dernier élément de ma_copie

ma_copie.pop()
print(ma_copie)
```

```python
print(ma_liste)
```

```python
# Copier une liste avec copy

ma_copie = ma_liste.copy()
print(ma_copie)
```

```python
ma_copie.pop()
print(ma_copie)
print(ma_liste)
```

```python
# Attention : la copie réalisée avec copie est superficielle

list1 = [1, 2, [3, 4]]
list2 = list1.copy()


list2[2][0] = 5

print(list1)
print(list2)
```

```python
# Copie profonde avec deepcopy

import copy

list1 = [1, 2, [3, 4]]
list2 = copy.deepcopy(list1)

list2[2][0] = 5

print(list1)
print(list2)
```

```python
# Calculer les carrés d'une liste et sauvegarder dans liste_carre

liste_carre = []

for i in range(1000_000):
  liste_carre.append(i**2)

liste_carre [:5]
```

```python
# Calculer les carrés avec une comprehension de liste

liste_carre = [i**2 for i in range(1000_000)]

print(liste_carre[:5])
```

```python
# Parcourir deux listes

liste1 = [1, 2, 3]
liste2 = ["a", "b", "c"]

for i, j in zip(liste1, liste2):
  print(i, j)
```

```python
# Parcourir une liste avec enumérate

for i, j in enumerate(liste2):
  print(i, j)
```

## Les dictionnaires

```python
# Définir un dictionnaire avec {}

mon_dict = {"Nom" : "Alice", "Age" : 38, "Ville" : "Lyon"}
print(mon_dict)
```

```python
# Accéder à la valeur de Age

mon_dict["Age"]
```

```python
# Modifier la valeur de l'âge

mon_dict["Age"] = 39
print(mon_dict)
```

```python
# Ajouter une clé profession

mon_dict["Profession"] = "Ingénieur"
print(mon_dict)
```

```python
# Récupérer les clés du dictionnaire

mon_obj = mon_dict.keys()
print(mon_obj)
```

```python
# Sauvegarder les clés d'un dictionnaire dans une liste
mes_cles = list(mon_dict.keys())
print(mes_cles)
```

```python
mes_cles[0]
```

```python
# Sauvegarder les valeurs d'un dictionnaire dans une liste
mes_valeurs = list(mon_dict.values())
print(mes_valeurs)
```

```python
# Créer un dictionnaire à partir de deux listes

cles = ["Nom", "Age", "Ville"]
valeurs = ["Bob", 42, "Paris"]

bob_dict = dict(zip(cles, valeurs))
print(bob_dict)
```
